# Visualization : Fermat distance - Coefficient of Variation

Estimation of the Fermat distance coefficient of variation. 
$$
\mathbb{CV} = \sqrt{\dfrac{\mathbb{V}[\hat{d}_{\mathbb{X}_n,\alpha}(x,y)]}{\mathbb{E}[\hat{d}_{\mathbb{X}_n,\alpha}(x,y)]^2}}.
$$


1. [$d$-dimensional cylinder of axis $[0,1]$ (along the first dimension) and radius 1](#part1)
2. [Moons dataset](#part2)

# Test of the **fermat_simu_cylinder** function

We can run a simple example for a simulation to see how this works:

```python
#No Harmonizer

n = 50 #Sample size
d = 2
all_alpha = np.linspace(1, 8, 29, endpoint=True) #Fermat dist parameter
n_simulations = 500 #Number of experiences

dataset = 'Uniform'
h = 0
harmonizer = [None,'Knn','Epsilon_radius','Delaunay'][h]
n_harmonizations = 1
#k = Best_param #Need to be optimized first
k = 20
#eps = Best_param #Need to be optimized first
eps = 0.25
seed = 1
```

```python
%%time
df = fermat_simu_cylinder(n=n, d=d, alphas=all_alpha, x_max=1, n_simulations=n_simulations, harmonizer=harmonizer, n_harmonizers=n_harmonizations, k=k, eps=eps, seed=seed)
df.to_csv('xp_result/CV_Cylinder_test.csv')
df
```

```python
df_combined = process_df_Eps(df)
df_combined
```

```python
fig, axes = plt.subplots(figsize=(8,4.5))

df_combined.plot('alpha', 'variation', ax=axes)
df_combined.plot('alpha', 'variation_lower', ax=axes)
df_combined.plot('alpha', 'variation_upper', ax=axes)
axes.set_ylabel('Coefficient of Variation')
axes.set_title(f"Uniform sample on a d-dimentional cylinder with axis=[0,1] and radius=1 \n d = {df_combined['d'][1]} - sample size = {df_combined['n'][1]} - harmonizer = {df_combined['harmonizer'][1]} - n_simulations = 50");
```
